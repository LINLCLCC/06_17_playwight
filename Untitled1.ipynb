## Pythagorean Theorem

For a right-angled triangle, the square of the hypotenuse (the side opposite the right angle) is equal to the sum of the squares of the other two sides. This is known as the Pythagorean theorem and is expressed as:

$a^2 + b^2 = c^2$

Where:
*   `a` and `b` are the lengths of the two shorter sides of the triangle.
*   `c` is the length of the hypotenuse.

```python
# Given the lengths of the two shorter sides (a and b)
a = 3
b = 4

# Calculate the square of the hypotenuse (c^2)
c_squared = a**2 + b**2

# Calculate the hypotenuse (c)
c = c_squared**0.5

print(f"The length of the hypotenuse is: {c}")
```

接下來，我們將計算您提供的數字序列的標準差。

```python
import statistics

# 您提供的數字序列
numbers = [3, 6, 9, 11, 17, 32]

# 計算標準差
# statistics.stdev() 函數計算樣本標準差
standard_deviation = statistics.stdev(numbers)

print(f"數字序列 {numbers} 的標準差是: {standard_deviation}")
```
